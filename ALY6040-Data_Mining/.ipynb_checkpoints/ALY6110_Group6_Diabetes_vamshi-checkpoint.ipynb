```python
df = pd.read_csv("./dataset_diabetes/dataset_diabetes/diabetic_data.csv")
dx=df.copy()
df
```

```python

drop=['medical_specialty', 'payer_code',  'patient_nbr','encounter_id','weight']
df.drop(columns=drop, inplace=True)
# Drop remaining rows with unknown values
df.drop(index=(df[df.isin(['?']).any(axis=1)].index), inplace=True)
# Let's check what's left
df.shape
df
```

```python
df.drop(columns=['admission_type_id', 'discharge_disposition_id', 'admission_source_id'], inplace=True)
```

```python
df["women"] = False
df.loc[df.gender == "Female", "women"] = True
df = df[df.gender != "Unknown/Invalid"]
df.drop(columns=['gender'], inplace=True)
```

```python
df = pd.get_dummies(data = df, columns = ["race"], prefix = "race", drop_first=False)
```

```python
regular = ['pioglitazone', 'rosiglitazone', 'insulin', 'glimepiride', 'glipizide',
                'glyburide','metformin', 'repaglinide'  ]
scarce = ["nateglinide", "chlorpropamide","troglitazone", "tolazamide", "examide",
             "citoglipton", "glyburide-metformin", "glipizide-metformin",
             "glimepiride-pioglitazone", "metformin-rosiglitazone", "metformin-pioglitazone", "acetohexamide", "tolbutamide",
             "acarbose", "miglitol"]

typesdrugs = regular + scarce
```

```python
for drug in regular :
    name = "t_" + drug
    df[name] = df[drug].isin(["Down", "Steady", "Up"])
```

```python
df.drop(columns=typesdrugs, inplace=True)
```

```python
def coding(col, codeDict):
  colCoded = pd.Series(col, copy=True)
  for key, value in codeDict.items():
    colCoded.replace(key, value, inplace=True)
  return colCoded
```

```python
readmit_encode_dict = {'NO':0, '<30':1, '>30':1}
df["readmitted"] = coding(df["readmitted"], readmit_encode_dict)
```

```python
df = pd.get_dummies(data = df, columns = ["readmitted"], prefix = "readmit", drop_first=False)
# I didn't use drop_first to be sure I was dropping readmit_NO
df.columns
```

```python
df = df.drop(["readmit_0"], axis = 1)
df.rename(columns = {'readmit_1': 'readmitted'}, inplace = True)
```

```python
df.loc[df.change == "Ch", "change"] = True
df.loc[df.change == "No", "change"] = False

df.loc[df.diabetesMed == "Yes", "diabetesMed"] = True
df.loc[df.diabetesMed == "No", "diabetesMed"] = False
```

```python
def editage (df) :
    return int(df.age[-4::].replace('-', '').replace(')', ''))

df["decade"] = df.apply(editage, axis = 1)
#var_quanti.append("age_num")
df = df.drop("age", axis = 1)
```

```python

dropped_Cols=[ 'diag_1', 'diag_2', 'diag_3']
df.drop(columns=dropped_Cols, inplace=True)
```

```python
df.drop(columns=['max_glu_serum','A1Cresult'], inplace=True)
```

## Random Forest

```python
X = df.drop(['readmitted'],axis=1)
y = df["readmitted"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
rf = RandomForestClassifier(n_estimators=500, random_state=1)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
```

```python
# variable (feature) importance plot
importances = rf.feature_importances_
std = np.std([tree.feature_importances_ for tree in rf.estimators_], axis=0)
rf_importance = pd.DataFrame({'feature': X_train.columns, 'importance': importances, 'std': std})
rf_importance = rf_importance.sort_values('importance',ascending=False)

pd.crosstab(pd.Series(y_test, name = 'Actual'), pd.Series(rf_pred, name = 'Predict'), margins = True)
```

```python
accuracy_rf = accuracy_score(y_test, rf_pred)
precision_rf = precision_score(y_test, rf_pred)
recall_rf = recall_score(y_test, rf_pred)
summary_rf = classificationSummary(y_test, rf.predict(X_test))

print("Accuracy is {0:.2f}".format(accuracy_rf))
print("Precision is {0:.2f}".format(precision_rf))
print("Recall is {0:.2f}".format(recall_rf))
```

```python
rf_importance
```

```python
plt.title('Feature Importances')
plt.barh(rf_importance['feature'],rf_importance['importance'], color='b', align='center')
plt.figure(figsize=(150,50))
plt.yticks(rf_importance['feature'])
plt.xlabel('Relative Importance')
plt.savefig("feature_importance.png")
plt.show()
```
